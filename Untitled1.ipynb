```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lattice_size = 22
num_monomers = 20
epsilon = 2.0
initial_temperature = 10
temperature_step = 0.5
final_temperature = 1
mc_steps = 10**6
last_50000_steps = 50000

# Function to calculate energy
def calculate_energy(lattice, positions):
    energy = 0
    for i in range(num_monomers):
        x0, y0 = positions[i]
        for dx, dy in [(-1,0), (1,0), (0,-1), (0,1)]:  # Nearest neighbors
            xn, yn = (x0 + dx) % lattice_size, (y0 + dy) % lattice_size
            if lattice[xn, yn] == 1 and (xn, yn) not in positions:
                energy += epsilon
    return energy / 2  # Each pair counted twice

# Monte Carlo move
def mc_move(lattice, positions, temperature):
    i = np.random.randint(num_monomers)  # Choose a random amino acid
    x0, y0 = positions[i]

    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # List of possible directions
    direction = directions[np.random.randint(4)]  # Randomly select one direction

    xn, yn = (x0 + direction[0]) % lattice_size, (y0 + direction[1]) % lattice_size

    if lattice[xn, yn] == 0:  # Check if the site is empty
        # Calculate energy difference
        positions[i] = (xn, yn)
        new_energy = calculate_energy(lattice, positions)
        positions[i] = (x0, y0)
        old_energy = calculate_energy(lattice, positions)

        delta_E = new_energy - old_energy

        # Move if energy is lower, or probabilistically if higher
        if delta_E < 0 or np.random.rand() < np.exp(-delta_E / temperature):
            lattice[x0, y0] = 0  # Free old site
            lattice[xn, yn] = 1  # Occupy new site
            positions[i] = (xn, yn)
            return True  # Move accepted

    return False  # Move rejected

# Initialize lattice and positions
def initialize_lattice():
    lattice = np.zeros((lattice_size, lattice_size), dtype=int)
    positions = [(i, lattice_size//2) for i in range(num_monomers)]
    for (x, y) in positions:
        lattice[x, y] = 1
    return lattice, positions

# Simulation function at different temperatures
def simulate_protein_folding(temperatures, mc_steps):
    avg_energies = []
    avg_end_to_end_lengths = []

    for T in temperatures:
        lattice, positions = initialize_lattice()
        energies = []
        end_to_end_lengths = []
        accepted_moves = 0

        for step in range(mc_steps):
            move_accepted = mc_move(lattice, positions, T)
            if move_accepted:
                accepted_moves += 1

            # Track only after a certain number of steps
            if step >= mc_steps - last_50000_steps:
                energies.append(calculate_energy(lattice, positions))
                end_to_end_lengths.append(np.linalg.norm(np.array(positions[0]) - np.array(positions[-1])))

            if step % 100000 == 0:  # Debug: print energy every 100,000 steps
                current_energy = calculate_energy(lattice, positions)
                print(f"Step: {step}, Temperature: {T}, Energy: {current_energy}, Accepted Moves: {accepted_moves}")

        # Average over the last 50,000 steps
        avg_energy = np.mean(energies)
        avg_end_to_end = np.mean(end_to_end_lengths)

        avg_energies.append(avg_energy)
        avg_end_to_end_lengths.append(avg_end_to_end)

    return avg_energies, avg_end_to_end_lengths

# Temperature range
temperatures = np.arange(initial_temperature, final_temperature - temperature_step, -temperature_step)

# Run the simulation over the temperature range
avg_energies, avg_end_to_end_lengths = simulate_protein_folding(temperatures, mc_steps)

# Plot the results
plt.figure(figsize=(12, 6))

# Average Energy plot
plt.subplot(1, 2, 1)
plt.plot(temperatures, avg_energies, marker='o', linestyle='-', color='b')
plt.title('Average Energy vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Average Energy')

# Average End-to-End Length plot
plt.subplot(1, 2, 2)
plt.plot(temperatures, avg_end_to_end_lengths, marker='o', linestyle='-', color='r')
plt.title('Average End-to-End Length vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Average End-to-End Length')

plt.tight_layout()
plt.show()
```
